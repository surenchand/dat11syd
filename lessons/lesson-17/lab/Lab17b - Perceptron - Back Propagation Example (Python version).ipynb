## First Forward Pass:



### A maths proof that you'll need to know before we continue...

First, you should set up the functions that you will need to use.

Remember that the sigmoid function looks like this:

\begin{equation}
s(x) = \frac{1}{1-e^{-x}}
\end{equation}

If we let **$~u = 1-e^{-x}~$**, so that **$~s = u^{-1}~$**, then the first derivative w.r.t $x$ is calculated like this, using the ***Chain Rule*** from circa Age-16 maths:

\begin{equation}
s'(x) = \frac{d\,s}{d\,x} = \frac{d s}{d u} \times \frac{d u}{d x}
\end{equation}

\begin{equation}
\frac{d\,s}{d\,u} = \frac{d ~u^{-1}}{d u} = -u^{-2} = -s^2
\end{equation}

\begin{equation}
\frac{d\,u}{d\,x} = e^{-x} = 1 - u
\end{equation}

\begin{equation}
s'(x) = \frac{d\,s}{d\,x} = -s^2 . (1-u) \\= -s . s(1-u) \\= -s . (s - su) \\= -s . (s - 1) \\= s(1-s)
\end{equation}

\begin{equation}
s'(x) = s(1-s)
\end{equation}
