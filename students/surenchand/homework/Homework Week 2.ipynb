```python
# Import the libraries required
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# this allows plots to appear directly in the notebook
%matplotlib inline
```

```python
plane_crash_data = pd.read_csv("~/workspace/dat11syd/students/surenchand/data/Airplane_Crashes_and_Fatalities_Since_1908.csv")
```

```python
plane_crash_data.head
```

```python
plane_crash_data.isnull().sum()
```

```python
plane_crash_data.dropna(axis=0,subset=['Operator'])
plane_crash_data.dropna(axis=0,subset=['Type'])
```

```python
pd.scatter_matrix(plane_crash_data, figsize=(15,15))
```

```python
plane_crash_data.plot(kind='scatter', x='Aboard', y='Fatalities')
```

```python
plane_crash_data.corr()
```

```python
lm = smf.ols(formula='Fatalities ~ Aboard', data=plane_crash_data).fit()
```

```python
lm.params
```

```python
lm.summary()
```

```python
import seaborn as sns
sns.set_style("darkgrid")

sns.lmplot(y='Fatalities', x='Aboard', data=plane_crash_data)
```

```python
x = plane_crash_data.Aboard
EstimatedFatalities = lm.predict(x)
```

```python
# Now, plot a line over the points that uses just the two points
plane_crash_data.plot(kind='scatter', x='EstimatedFatalities', y='Fatalities')
# this code overlays a straight line between the the coordinates created by X_new and preds
#plt.plot(Fatalities, EstimatedFatalities, c='red', linewidth=2)
```
